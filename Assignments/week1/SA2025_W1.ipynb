### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
if not df_graded.empty and all(col in df_graded.columns for col in ['horsepower', 'weight', 'origin', 'mpg']):
    df_plot_scatter = df_graded.copy()

    # Ensure relevant columns are numeric
    df_plot_scatter['horsepower'] = pd.to_numeric(df_plot_scatter['horsepower'], errors='coerce')
    df_plot_scatter['weight'] = pd.to_numeric(df_plot_scatter['weight'], errors='coerce')
    df_plot_scatter['mpg'] = pd.to_numeric(df_plot_scatter['mpg'], errors='coerce')

    # Prepare 'origin' column for hue_order. It must match the strings in hue_order_list.
    # If 'origin' is numeric (e.g., 1, 2, 3), map it to strings ('usa', 'europe', 'japan').
    # The hue_order expects these specific strings.
    hue_order_list = ['japan', 'europe', 'usa']

    # Create a new column for mapped origin strings to avoid modifying original 'origin' if not desired
    df_plot_scatter['origin_mapped'] = df_plot_scatter['origin']

    if pd.api.types.is_numeric_dtype(df_plot_scatter['origin']):
        # Assuming: 1 maps to 'usa', 2 to 'europe', 3 to 'japan' to match hue_order_list.
        # This mapping is crucial and must be correct for your dataset.
        origin_mapping = {1: 'usa', 2: 'europe', 3: 'japan'}
        df_plot_scatter['origin_mapped'] = df_plot_scatter['origin'].map(origin_mapping)
        print("Mapped numeric 'origin' to strings for plotting.")
    elif pd.api.types.is_string_dtype(df_plot_scatter['origin']):
        # If already string, ensure they are lowercase to match hue_order_list.
        df_plot_scatter['origin_mapped'] = df_plot_scatter['origin'].str.lower()
        print("Converted string 'origin' to lowercase for plotting.")
    else:
        print("Warning: 'origin' column is not clearly numeric or string. Plot might not work as expected with hue_order.")
        # Attempt to convert to string and lowercase as a fallback
        df_plot_scatter['origin_mapped'] = df_plot_scatter['origin'].astype(str).str.lower()


    # Drop rows where any of the critical plotting columns are NaN
    df_plot_scatter.dropna(subset=['horsepower', 'weight', 'mpg', 'origin_mapped'], inplace=True)

    # Filter out origins not in hue_order_list to prevent issues with hue_order
    df_plot_scatter = df_plot_scatter[df_plot_scatter['origin_mapped'].isin(hue_order_list)]


    if not df_plot_scatter.empty:
        plt.figure(figsize=(14, 9))
        scatter_plot = sns.scatterplot(
            x='horsepower',
            y='weight',
            hue='origin_mapped',
            size='mpg',
            sizes=(30, 250),
            alpha=0.7,
            data=df_plot_scatter,
            hue_order=hue_order_list
        )
        plt.title('Car Characteristics: Horsepower vs. Weight by Origin', fontsize=18)
        plt.xlabel('Horsepower', fontsize=15)
        plt.ylabel('Weight (lbs)', fontsize=15)


        handles, labels = scatter_plot.get_legend_handles_labels()

        plt.legend(title='Origin & MPG (Size)', loc='best', frameon=True, fontsize=10, title_fontsize=12)

        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
    else:
        print("Not enough valid data to create the scatter plot after cleaning and mapping 'origin'.")
        print(f"Check unique values in 'origin_mapped' after processing: {df_graded['origin'].map({1:'usa',2:'europe',3:'japan'}).dropna().unique() if pd.api.types.is_numeric_dtype(df_graded['origin']) else df_graded['origin'].str.lower().dropna().unique()}")


elif not df_graded.empty:
    missing_cols = [col for col in ['horsepower', 'weight', 'origin', 'mpg'] if col not in df_graded.columns]
    print(f"Missing required columns in df_graded for Ungraded Question 3: {', '.join(missing_cols)}")
else:
    print("DataFrame df_graded is empty.")
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python

if not df_graded.empty and all(col in df_graded.columns for col in ['name', 'model_year', 'mpg']):

    df_consistent_analysis = df_graded.copy()
    df_consistent_analysis['mpg'] = pd.to_numeric(df_consistent_analysis['mpg'], errors='coerce')
    df_consistent_analysis['model_year'] = pd.to_numeric(df_consistent_analysis['model_year'], errors='coerce')


    df_consistent_analysis.dropna(subset=['name', 'model_year', 'mpg'], inplace=True)

    car_appearances = df_consistent_analysis.groupby('name')['model_year'].nunique()
    multi_year_cars_names = car_appearances[car_appearances > 1].index.tolist()

    if multi_year_cars_names:

        df_multi_year_cars = df_consistent_analysis[df_consistent_analysis['name'].isin(multi_year_cars_names)].copy()


        consistent_cars_stats = df_multi_year_cars.groupby('name').agg(
            mpg_std=('mpg', 'std'),
            mpg_mean=('mpg', 'mean'),
            num_unique_years=('model_year', 'nunique')
        ).reset_index()


        consistent_cars_final = consistent_cars_stats[
            (consistent_cars_stats['mpg_std'] < 1.0)
        ].copy()

        if not consistent_cars_final.empty:

            consistent_cars_final.sort_values(
                by=['num_unique_years', 'mpg_mean'],
                ascending=[False, False],
                inplace=True
            )

            print("Consistent Car Models (Produced over multiple years with MPG std < 1.0):")

            consistent_cars_final.rename(columns={'num_unique_years': 'num_model_years'}, inplace=True)
            print(consistent_cars_final[['name', 'num_model_years', 'mpg_mean', 'mpg_std']])
        else:
            print("No car models found that meet the 'consistent' criteria (produced over multiple years with std(mpg) < 1.0).")
    else:
        print("No car models found that were produced in more than one model year in the cleaned dataset.")

elif not df_graded.empty:
    missing_cols = [col for col in ['name', 'model_year', 'mpg'] if col not in df_graded.columns]
    print(f"Missing required columns in df_graded for Ungraded Question 4: {', '.join(missing_cols)}")
else:
    print("DataFrame df_graded is empty.")
```
